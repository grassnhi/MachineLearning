Inter Quantile Range (IQR)

‚ñ∂ Ngo√†i ra, ta c√≥ th·ªÉ d√πng IQR (kho·∫£ng t·ª© ph√¢n v·ªã) ƒë·ªÉ x√°c ƒë·ªãnh ph·∫°m vi ph·∫ßn l·ªõn d·ªØ li·ªáu t·∫≠p trung `[lower,upper]` v√† x·ª≠ l√Ω c√°c ngo·∫°i l·ªá n·∫±m b√™n ngo√†i ƒëo·∫°n n√†y b·∫±ng ph∆∞∆°ng ph√°p **Quantile based flooring and capping**. C√°c gi√° tr·ªã n·∫±m ngo√†i ƒëo·∫°n `[lower,upper]` s·∫Ω ƒë∆∞·ª£c g√°n cho gi√° tr·ªã c·ªßa upper n·∫øu gi√° tr·ªã c·ªßa n√≥ > upper v√† lower n·∫øu gi√° tr·ªã c·ªßa n√≥ < lower.

```python
from typing import Tuple
from sklearn.base import BaseEstimator, TransformerMixin

def find_boxplot_boundaries(
    col: pd.Series, whisker_coeff: float = 1.5
) -> Tuple[float, float]:
    """Findx minimum and maximum in boxplot.

    Args:
        col: a pandas serires of input.
        whisker_coeff: whisker coefficient in box plot
    """
    Q1 = col.quantile(0.25)
    Q3 = col.quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - whisker_coeff * IQR
    upper = Q3 + whisker_coeff * IQR
    return lower, upper


class BoxplotOutlierClipper(BaseEstimator, TransformerMixin):
    def __init__(self, whisker_coeff: float = 1.5):
        self.whisker = whisker_coeff
        self.lower = None
        self.upper = None

    def fit(self, X: pd.Series):
        self.lower, self.upper = find_boxplot_boundaries(X, self.whisker)
        return self

    def transform(self, X):
        return X.clip(self.lower, self.upper)
```

```python
clipped_absences = BoxplotOutlierClipper().fit_transform(df["absences"])

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(15, 5))
clipped_absences.hist(bins=40, ax=axes[0])
clipped_absences.to_frame().boxplot(ax=axes[1], vert=False);
```

### 4.2. Feature encoding (m√£ h√≥a ƒë·∫∑c tr∆∞ng)


‚öì Feature encoding l√† g√¨? Th∆∞·ªùng ƒë∆∞·ª£c √°p d·ª•ng cho lo·∫°i thu·ªôc t√≠nh n√†o?

- Lo·∫°i thu·ªôc t√≠nh th∆∞·ªùng ƒë∆∞·ª£c √°p d·ª•ng: d·ªØ li·ªáu h·∫°ng m·ª•c (categorical data)

‚öì T·∫°i sao feature encoding quan tr·ªçng trong qu√° tr√¨nh data preparation? 

- V√¨ x·ª≠ l√Ω to√°n h·ªçc, ƒë√†o t·∫°o m√¥ h√¨nh ƒë·ªÅu ph·∫£i thao t√°c v·ªõi c√°c con s·ªë. Ta ph·∫£i chuy·ªÉn ƒë·ªïi c√°c d·ªØ li·ªáu t·ª´ d·∫°ng ch·ªØ (text) sang d·∫°ng s·ªë (numbers) ƒë·ªÉ c√≥ th·ªÉ √°p d·ª•ng c√°c m√¥ h√¨nh h·ªçc m√°y/thu·∫≠t to√°n.

‚öì C√°c ki·ªÉu categorical feature:
- Binary
- Ordinary
- Nominal 
- Numeric

‚öì M·ªôt s·ªë k·ªπ thu·∫≠t encoding th∆∞·ªùng d√πng:
- Label Encoding (th∆∞·ªùng ƒë∆∞·ª£c d√πng cho binary ho·∫∑c nominal)
- One-hot Encoding (th∆∞∆°ng d√πng cho nominal)
- Ordinal Encoding
- Ngo√†i ra, c√≤n c√≥ Frequency Encoding, Binary Encoding, Mean Encoding, Hash Encoding‚Ä¶

‚ñ∂ **B∆∞·ªõc 1**: Feature Encoding cho c√°c ƒë·∫∑c tr∆∞ng d·∫°ng binary (yes/no)

Ta s·ª≠ d·ª•ng b·ªô m√£ h√≥a `LabelEncoder` c·ªßa th∆∞ vi·ªán ti·ªÅn x·ª≠ l√Ω t·ª´ sklearn. 

```python
# feature encoding for binary features (13 features)
binary_feature = ['School','sex','address','family_size','parents_status','school_support','family_support','paid_classes','activities','nursery','desire_higher_edu','internet','romantic']
from sklearn.preprocessing import LabelEncoder  
le = LabelEncoder()

for column in binary_feature: 
  df[column +"_encoded"] = le.fit_transform(df[column])
```

‚ñ∂ **B∆∞·ªõc 2**: Feature Encoding cho c√°c ƒë·∫∑c tr∆∞ng d·∫°ng nominal. 

Ta s·∫Ω s·ª≠ d·ª•ng ph∆∞∆°ng ph√°p m√£ h√≥a ***One-hot Encoding*** b·∫±ng method `get_dummies` c·ªßa **Pandas**. ƒê·ªëi v·ªõi m·ªói c·ªôt ƒë·∫∑c tr∆∞ng d·∫°ng nominal, s·ªë l∆∞·ª£ng c·ªôt m·ªõi ƒë∆∞·ª£c t·∫°o ra s·∫Ω b·∫±ng s·ªë l∆∞·ª£ng th·ªÉ lo·∫°i c·ªßa ƒë·∫∑c tr∆∞ng ƒë√≥. 

```python
# feature encoding for nominal features (3 features)
nominal_feature = ['mother_job','father_job','guardian']

for column in nominal_feature: 
  df = pd.get_dummies(df, prefix = [column], columns = [column])
```

### 5.3. Relationship between Final Grade (label) & input attributes




‚öì M·ª•c ƒë√≠ch:

‚ñ∂ ƒê·∫ßu ti√™n, ch√∫ng ta xem x√©t s·ª± ·∫£nh h∆∞·ªüng c·ªßa romantic relationship ƒë·∫øn k·∫øt qu·∫£ h·ªçc t·∫≠p c·ªßa c√°c sinh vi√™n

```python
# romantic status
perc = (lambda col: col/col.sum())
index = ['poor','fair','good']
# compute a cross tabulation (frequency table) of final_grade & romantic 
romance_tab1 = pd.crosstab(index=df_main.final_grade, columns=df_main.romantic)
# convert to log value because we only need to compare quantity between pairs of (final_grade, romantic)
romance_tab = np.log(romance_tab1) 
# convert to percentage for each col and reindex 
romance_perc = romance_tab.apply(perc).reindex(index)
```

Sinh vi√™n trong m·ªëi quan h·ªá y√™u ƒë∆∞∆°ng c√≥ xu h∆∞·ªõng ƒë·∫°t k·∫øt qu·∫£ h·ªçc t·∫≠p k√©m (poor final grade), ng∆∞·ª£c l·∫°i nh·ªØng sinh vi√™n kh√¥ng trong m·ªëi quan h·ªá y√™u ƒë∆∞∆°ng c√≥ xu h∆∞·ªõng ƒë·∫°t k·∫øt qu·∫£ cao trong h·ªçc t·∫≠p (good score).


‚öì**Gi·ªõi thi·ªáu v·ªÅ ki·ªÉm ƒë·ªãnh chi b√¨nh ph∆∞∆°ng:**

- D√πng ƒë·ªÉ ki·ªÉm tra xem c√≥ m·ªëi quan h·ªá gi·ªØa 2 bi·∫øn (2 y·∫øu t·ªë) ƒëang x√©t hay kh√¥ng. T·ªïng qu√°t h∆°n, chi-square test ki·ªÉm tra li·ªáu ph√¢n ph·ªëi c·ªßa categorical variables c√≥ kh√°c so v·ªõi bi·∫øn c√≤n l·∫°i. 
- Nhu c·∫ßu ki·ªÉm tra m·ªëi li√™n h·ªá gi·ªØa 2 bi·∫øn xu·∫•t hi·ªán nhi·ªÅu trong c√°c ph∆∞∆°ng ph√°p ph√¢n t√≠ch nh∆∞ h·ªìi quy Logistics, Survival analysis,... 
- C√¥ng th·ª©c: 

chi-square-formula.jpg

V·ªõi O l√† gi√° tr·ªã quan s√°t, E l√† gi√° tr·ªã mong ƒë·ª£i v√† c l√† b·∫≠c t·ª± do.

** C√°ch t√≠nh gi√° tr·ªã mong ƒë·ª£i:
https://giamcanherbalthin.com/chi-square-la-gi/ 
- N·∫øu gi√° tr·ªã chi-square nh·ªè, c√≥ m·ªëi t∆∞∆°ng quan cao gi·ªØa 2 t·∫≠p d·ªØ li·ªáu.
- N·∫øu gi√° tr·ªã chi-square ƒë·ªß l·ªõn (>10), th√¨ kh·∫£ nƒÉng s·∫Ω kh√¥ng c√≥ m·ªëi quan h·ªá nhi·ªÅu gi·ªØa 2 t·∫≠p d·ªØ li·ªáu.

** X√°c ƒë·ªãnh m·ªëi t∆∞∆°ng quan d·ª±a v√†o gi·∫£ thuy·∫øt H0 v√† p-value:  
- p-value: x√°c su·∫•t m·∫Øc ph·∫£i sai l·∫ßm n·∫øu ta b√°c b·ªè gi·∫£ thi·∫øt H0 (2 ti√™u chu·∫©n ph√¢n lo·∫°i ƒë·ªôc l·∫≠p v·ªõi nhau). N·∫øu p-value c√†ng b√©, kh·∫£ nƒÉng H0 ƒë∆∞·ª£c b√°c b·ªè c√†ng cao, t·ª©c l√† 2 bi·∫øn c√≥ m·ªëi t∆∞∆°ng quan v·ªõi nhau. ƒê·ªÉ x√°c ƒë·ªãnh ƒë∆∞·ª£c p-value, ta c·∫ßn c√≥ gi√° tr·ªã b·∫≠c t·ª± do v√† ch·ªâ s·ªë alpha.
- Gi√° tr·ªã p-value c√†ng nh·ªè (p-value < Œ±) th√¨ gi·∫£ thuy·∫øt kh√¥ng c√†ng n√™n b·ªã b√°c b·ªè v√† gi·∫£ thuy·∫øt thay th·∫ø c√†ng ƒë√°ng tin c·∫≠y. 
- Gi√° tr·ªã p-value c√†ng l·ªõn (p-value> Œ±) th√¨ c∆° s·ªü ƒë·ªÉ b√°c b·ªè gi·∫£ thuy·∫øt kh√¥ng c√†ng y·∫øu v√† ki·ªÉm ƒë·ªãnh kh√¥ng c√≥ k·∫øt lu·∫≠n. 


```python
# chi-square test result
import statsmodels.api as sm
# convert contingency table into appropriate format to apply api of statsmodel library
romance_table = sm.stats.Table(romance_tab1)
# conduct chi-square test 
romance_rslt = romance_table.test_nominal_association()
romance_rslt.pvalue
```

- Kh√¥ng c√≥ s·ª± kh√°c bi·ªát l·ªõn gi·ªØa t·ª∑ l·ªá h·ªçc sinh b·ªã ƒëi·ªÉm k√©m khi xem x√©t theo mother education level.

Ti·∫øp theo, nh√≥m s·∫Ω s·ª≠ d·ª•ng m·ªôt m√¥ h√¨nh h·ªìi quy OLS ƒë∆°n gi·∫£n ƒë·ªÉ ƒë√°nh gi√° s·ª± ph·ª• thu·ªôc c·ªßa ƒëi·ªÉm s·ªë cu·ªëi c√πng c·ªßa h·ªçc sinh v√†o tr√¨nh ƒë·ªô h·ªçc v·∫•n c·ªßa cha v√† m·∫π.

**OLS regression model (M√¥ h√¨nh h·ªìi quy OLS)** 

https://stataguide.wordpress.com/2020/04/13/mo-hinh-hoi-quy-ols/
- OLS - Ordinary Least Square (pp h·ªìi quy b√¨nh ph∆∞∆°ng t·ªëi thi·ªÉu) l√† ph∆∞∆°ng ph√°p h·ªìi quy ƒë∆∞·ª£c s·ª≠ d·ª•ng ph·ªï bi·∫øn nh·∫•t trong nghi√™n c·ª©u.

- Y v√† X ƒë∆∞·ª£c s·ª≠ d·ª•ng ƒë·ªÉ th·ªÉ hi·ªán cho t·∫•t c·∫£ quan s√°t c·ªßa m·ªôt t·ªïng th·ªÉ (population), c√≤n y v√† x th·ªÉ hi·ªán cho c√°c quan s√°t trong m·∫´u nghi√™n c·ª©u ƒë∆∞·ª£c ch·ªçn (sample). Khi ƒë√≥, c√°c h·ªá s·ªë alpha v√† beta ƒë∆∞·ª£c k√Ω hi·ªáu d·∫•u m≈© - c√°c gi√° tr·ªã ∆∞·ªõc l∆∞·ª£ng. 
- OLS s·∫Ω l·ª±a ch·ªçn c√°c h·ªá s·ªë h·ªìi quy alpha v√† beta sao cho t·ªïng b√¨nh ph∆∞∆°ng sai s·ªë c·ªßa m√¥ h√¨nh ∆∞·ªõc l∆∞·ª£ng l√† nh·ªè nh·∫•t.

-  C√≥ 3 v·∫•n ƒë·ªÅ m√† ch√∫ng ta c·∫ßn quan t√¢m ƒë·∫ßu ti√™n: ƒë√≥ l√† h·ªá s·ªë h·ªìi quy c√≥ √Ω nghƒ©a th·ªëng k√™ kh√¥ng, m√¥ h√¨nh c√≥ √Ω nghƒ©a kh√¥ng v√† m·ª©c ƒë·ªô gi·∫£i th√≠ch c·ªßa m√¥ h√¨nh nh∆∞ th·∫ø n√†o?

```python
# use OLS to see coefficients
X_edu = df[['mother_education','father_education']]
y_edu = df.final_score
edu = sm.OLS(y_edu, X_edu)
results_edu = edu.fit()
results_edu.summary()
```

- B1: ki·ªÉm ƒë·ªãnh gi·∫£ thuy·∫øt h·ªá s·ªë h·ªìi quy. M·ª•c ti√™u l√† b√°c b·ªè gi·∫£ thuy·∫øt null (beta = 0). ƒê·ªÉ th·ª±c hi·ªán ki·ªÉm ƒë·ªãnh n√†y, ta xem x√©t gi√° tr·ªã p-value t∆∞∆°ng ·ª©ng. Quan s√°t t√≥m t·∫Øt k·∫øt qu·∫£, ta th·∫•y c√°c bi·∫øn father_education v√† mother_education ƒë·ªÅu c√≥ ·∫£nh h∆∞·ªüng t√≠ch c·ª±c l√™n bi·∫øn ph·ª• thu·ªôc final_score.
- B2: ki·ªÉm ƒë·ªãnh m√¥ h√¨nh hay c√≤n g·ªçi l√† ki·ªÉm ƒë·ªãnh F (Gi·∫£ thuy·∫øt cho ki·ªÉm ƒë·ªãnh n√†y l√† t·∫•t c·∫£ c√°c h·ªá s·ªë h·ªìi quy ƒë·ªìng th·ªùi b·∫±ng 0). D√πng th·ªëng k√™ F.
- B3: ki·ªÉm tra gi√° tr·ªã c·ªßa R-square. R2 th·ªÉ hi·ªán cho % bi·∫øn thi√™n c·ªßa bi·∫øn ph·ª• thu·ªôc ƒë∆∞·ª£c gi·∫£i th√≠ch b·ªüi m√¥ h√¨nh. 

```python
# use GLS to see coefficients
edu = sm.GLS(y_edu, X_edu)
results_edu_wls = edu.fit()
results_edu_wls.summary()
```

**Swarm plot**:
- Swarm plot l√† m·ªôt d·∫°ng bi·ªÉu ƒë·ªì bi·ªÉu di·ªÖn ph√¢n ph·ªëi v√† th·ªÉ hi·ªán s·ª± ph√¢n ph·ªëi d·ªØ li·ªáu d∆∞·ªõi d·∫°ng c√°c ƒëi·ªÉm.
