## Combining

The simplest combining method is concat(). 

Given a list of elements, this function will smush those elements together along an axis.

```python
import pandas as pd
```

```python
canadian_youtube = pd.read_csv("E:/Coding/MachineLearning/CSVFile/CAvideos.csv")
british_youtube = pd.read_csv("E:/Coding/MachineLearning/CSVFile/GBvideos.csv")

pd.concat([canadian_youtube, british_youtube])
```

```python
left = canadian_youtube.set_index(['title', 'trending_date'])
right = british_youtube.set_index(['title', 'trending_date'])

left.join(right, lsuffix='_CAN', rsuffix='_UK')
```
